```python
# Here we assume that we have a client coming to us asking
# for an automated Rental Property Calculator. Our client's
# name is Brandon from a company called "Bigger Pockets". Below,
# you will find a video of what Brandon usually does to calculate
# his Rental Property ROI.
# Using Visual Studio Code/Jupyter Notebook, and Object Oriented Programming
# create a program that will calculate the Return on Investment(ROI)
# for a rental property.

# 4 Square method:

# 1. Income to calculate:
# questions to ask: expected rental income?
# Any Expected Laundry or storage income?
# Any other miscellaneous income?
# Show Total Monthly and Total Annual Income

# 2. Expenses to calculate:
# Down payment:
# Taxes to be paid:
# Insurance to be paid
# Show Total monthly and total annual mortgage costs

# Landlord to pay Utilities? if yes:
# Electric, water, sewage, garbage, gas:
# Hoa fees?
# lawn and snow?
# Vacancy?
# Repairs?
# CapEx?
# Property Management?
# Show Total monthly and total annual expenses

# 3. Cash Flow:
# Total Annual Income - Total Annual Expenses = Total Cash flow
# Show monthly and Annual


# 4. Return on Investment:
# Down Payment?
# Closing Costs?
# Repair Money?
# Any other expected costs
# Sum of above = Total initial Investment
# Total annual Cash flow / Total initial investment x 100 = Rate of return
# Show Total Annual income, Total annual expenses, Total initial investment and rate of return.
```

```python
from IPython.display import clear_output

class RentalRoi:

    def __init__(self, mon_inc=0, mon_exp=0, an_inc=0,
                 an_exp=0, an_cashflow=0, mon_cashflow=0):
        self.mon_inc = mon_inc
        self.mon_exp = mon_exp
        self.an_inc = an_inc
        self.an_exp = an_exp
        self.an_cashflow = an_cashflow
        self.mon_cashflow = mon_cashflow

    def cal_income(self):
        while True:
            print(f'Your current rental income is ${self.mon_inc}.')
            exp_inc = input("Let's figure out your expected income from this property. "
                            "\nHow much rental income do you expect to earn per month?\nType "
                            "the number amount or type 'a' to restart income, "
                            "type '0' if the current income is ok, "
                            "or type 'r' to return to main menu: ")
            clear_output()
            if exp_inc.lower() == 'r':
                clear_output()
                return
            elif exp_inc == 'a':
                self.mon_inc *= 0
                self.an_inc *= 0
                clear_output()
                continue
            elif exp_inc.isnumeric():
                exp_inc = int(exp_inc)
                self.mon_inc += exp_inc
                self.an_inc = self.mon_inc * 12
            else:
                clear_output()
                print("Please enter a valid number. ")
                continue
            print(f'Your current monthly income is ${self.mon_inc}.')
            other_inc_check = input("Ok Now please type the amount of any extra income \nyou might expect. "
                                    " You can also type 'f' to finish with income \n"
                                    "and move to the next section (expenses) or type 'r' "
                                    "to return to the main menu: ")
            clear_output()
            if other_inc_check.lower() == 'r':
                clear_output()
                return
            elif other_inc_check.lower() == 'f':
                self.cal_expenses()
            elif other_inc_check.isnumeric():
                other_inc_check = int(other_inc_check)
                self.mon_inc += other_inc_check
                self.an_inc = self.mon_inc * 12
                print(f'Ok your total monthly income is ${self.mon_inc} \nand your total annual income is ${self.an_inc}.')
            else:
                print("Please provide a valid input. Let's start 'income' over so we are sure of what we are doing.")
                continue
            return

    def cal_expenses(self):
        while True:
            print(f'Your current monthly expenses are ${self.mon_exp}')
            mortgage = input("Let's figure out your expenses by starting with your mortgage. "
                             "What's the monthly mortgage amount? \n"
                             " We'll add taxes and insurance to it in a moment. \nType the number amount or "
                             "type 'a' to restart expenses, type '0' if the current \n "
                             " income is ok, or type 'r' to return to the main menu: ")
            clear_output()
            if mortgage.lower() == 'r':
                clear_output()
                return
            elif mortgage == 'a':
                self.mon_exp *= 0
                self.an_exp *= 0
                continue
            elif mortgage.isnumeric():
                mortgage = int(mortgage)
                self.mon_exp += mortgage
                self.mon_cashflow = self.mon_inc - self.mon_exp
                self.an_cashflow = self.an_inc - self.an_exp
            else:
                clear_output()
                print("Please enter a valid number.")
                continue
            tax = input("please enter the tax amount you will pay per month on this property:")
            clear_output()
            if tax.isnumeric():
                tax = int(tax)
                self.mon_exp += tax
                self.mon_cashflow = self.mon_inc - self.mon_exp
                self.an_cashflow = self.an_inc - self.an_exp
            else:
                clear_output()
                print("Please enter a valid number. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            insur = input("Please enter the monthly cost of insuring this property:")
            clear_output()
            if insur.isnumeric():
                insur = int(insur)
                self.mon_exp += insur
                self.mon_cashflow = self.mon_inc - self.mon_exp
                self.an_cashflow = self.an_inc - self.an_exp
            else:
                clear_output()
                print("Please enter a valid number. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            self.an_exp = self.mon_exp * 12
            self.mon_cashflow = self.mon_inc - self.mon_exp
            self.an_cashflow = self.an_inc - self.an_exp
            h_o_a = input(f"Ok your annual mortgage is ${self.an_exp} and your monthly mortgage is ${self.mon_exp}.\n "
                          f"Now let's check other expenses. Please enter how much you\n "
                          f"will pay in monthly HOA fees, (type 0 if there aren't any fees).\n "
                          " Type 'a' to reset expenses to '0', "
                          "or type 'r' to return to the main menu: ")
            clear_output()
            if h_o_a.lower() == 'r':
                clear_output()
                return
            elif mortgage == 'a':
                self.mon_exp *= 0
                self.an_exp *= 0
                continue
            elif h_o_a.isnumeric():
                h_o_a = int(h_o_a)
                self.mon_exp += h_o_a
                self.an_exp = self.mon_exp * 12
            else:
                print("Please provide a valid input. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            util = input("Will you be paying the utilities or the tenant? If you are paying \n"
                         "them please type the total estimated monthly "
                         "cost\n(remember to consider gas, water, electricity, garbage collection etc)\n"
                         "or you can type 'r' to return to the main menu: ")
            clear_output()
            if util == 'r':
                clear_output()
                return
            elif util.isnumeric():
                util = int(util)
                self.mon_exp += util
                self.an_exp = self.mon_exp * 12
            else:
                clear_output()
                print("Please enter a valid input. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            vac_repair = input("Please enter how much you want to put aside each month for vacancy\n"
                               "(when your property sits empty) and repairs (broken windows, holes in the wall, "
                               "mold etc).\nI suggest setting aside $200 - $300 to be safe: ")
            clear_output()
            if vac_repair.isnumeric():
                vac_repair = int(vac_repair)
                self.mon_exp += vac_repair
                self.an_exp = self.mon_exp * 12
            else:
                clear_output()
                print("Please enter a valid number. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            prop_management = input(
                "Will you manage hire a property manager? Please enter the expected monthly cost\n (type "
                "'0' if you will manage the property yourself.\n "
                "You can also type 'r' to return to the main menu): ")
            clear_output()
            if prop_management.lower() == 'r':
                clear_output()
                return
            elif prop_management.isnumeric():
                prop_management = int(prop_management)
                self.mon_exp += prop_management
                self.an_exp = self.mon_exp * 12
            else:
                clear_output()
                print("Please provide a valid input. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            cap_ex = input("Please enter how much you will set aside per month for big expenditures\n"
                           " (fixing the roof etc) This is important, it will help protect\n "
                           "your investment in case of emergencies..I suggest setting aside $50 for this: ")
            clear_output()
            if cap_ex.isnumeric():
                cap_ex = int(cap_ex)
                self.mon_exp += cap_ex
                self.an_exp = self.mon_exp * 12
            else:
                clear_output()
                print("Please enter a number. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            misc_exp = input("Ok, now let's add any final expenses. If you plan to provide a lawn or snow service,\n"
                             "or can think of anything else you need to pay for,\n "
                             "please enter the monthly amount here: type '0' if you don't have any other expenses: ")
            clear_output()
            if misc_exp.isnumeric():
                misc_exp = int(misc_exp)
                self.mon_exp += misc_exp
                self.an_exp = self.mon_exp * 12
            else:
                clear_output()
                print("Please enter a number. Let's start 'expenses' over so we are sure of what we are doing.")
                continue
            print(f"Ok your total monthly expenses are ${self.mon_exp} and your total annual "
                  f"expenses are ${self.an_exp}.")
            return

    def cal_cashflow(self):
            self.mon_cashflow = self.mon_inc - self.mon_exp
            self.an_cashflow = self.an_inc - self.an_exp
            if self.mon_inc == 0 or self.mon_exp == 0:
                print(f'You must calculate both your income and expenses first!')
                return
            else:
                print(f'Current monthly income: ${self.mon_inc}')
                print(f'Current annual income: ${self.an_inc}')
                print(f'Current monthly expenses: ${self.mon_exp}')
                print(f'Current annual expenses: ${self.an_exp}')
                print(f'Your total monthly cash flow is {self.mon_cashflow} and your total '
                      f"annual cash flow is ${self.an_cashflow}.")
                
                return

    def cal_r_o_i(self):
        while True:
        
            if self.mon_inc == 0 or self.mon_exp == 0:
                print(f'You must calculate both your income and expenses first!')
                return
            down_p = input("Almost done! Only 3 short steps! Please enter your down payment (enter the purchase "
                           "price\n if you are paying cash upfront and won't have a mortgage):  ")
            clear_output()
            if down_p.isnumeric():
                down_p = int(down_p)
            else:
                clear_output()
                print("please enter a valid number")
                continue
            clos_costs = input("Please enter your expected closing costs: ")
            clear_output()
            if clos_costs.isnumeric():
                clos_costs = int(clos_costs)
            else:
                clear_output()
                print("Please enter a valid number.")
                continue
            repairs_and_more = input("Please enter how much you think you need to spend on repairs,\n or any "
                                     "other initial costs you think you might have: ")
            clear_output()
            if repairs_and_more.isnumeric():
                repairs_and_more = int(repairs_and_more)
            else:
                clear_output()
                print("Please enter a valid number.")
                continue
            initial_inv = down_p + clos_costs + repairs_and_more
            r_o_i_1 = (self.an_cashflow / initial_inv) * 100
            r_o_i = round(r_o_i_1, 2)
            print(f"Your initial investment is ${initial_inv} and your total annual "
                  f"cashflow is ${self.an_cashflow}.\n Your return on investment for this property is {r_o_i}% Thank "
                  f"you for using your Bigger Pockets ROI Calculator.")
            return

    def run_rental(self):
        print("Welcome to your Bigger Pockets Rental ROI calculator! \nPlease follow steps (1-4) to find your rental return of investment!")
        while True:
            user = input("Step 1: type 'i' to calculate income. \n"
                         "Step 2: type 'e' to calculate expenses.\n"
                         "Step 3: type 'cf' to see your cash flow.\n"
                         "Step 4:'roi' to see your return of investment.\n "
                         "(or type 'quit' to quit): ")
            if user == 'i':
                clear_output()
                self.cal_income()          
            elif user == 'e':
                clear_output()
                self.cal_expenses()
            elif user == 'cf':
                clear_output()
                self.cal_cashflow()
            elif user == 'roi':
                clear_output()
                self.cal_r_o_i()
            elif user == 'quit':
                return
            else:
                clear_output()
                print("That's not a valid response. Please try again")


rental_roi = RentalRoi()

rental_roi.run_rental()
```
